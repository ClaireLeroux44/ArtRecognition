# Images resizing

```python
import os
import matplotlib.pyplot as plt
from skimage.transform import resize
from IPython.display import display
from PIL import Image, ImageFile

ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
path0 = r'C:\Users\pitip\code\ClaireLeroux44\ArtRecognition'
path_data = r'raw_data/train_1'
path_resample_224 = r'raw_data/resample224'
path_resample_40 = r'raw_data/resample40'
```

```python
img_list = os.listdir(os.path.join(path0, path_data))
img_list[:10]
```

```python
im = plt.imread(os.path.join(path0, path_data, img_list[0]))
print(im.shape)
plt.imshow(im);
```

```python
image_224 = resize(im, (224, 224), anti_aliasing=False)
image_40 = resize(im, (40, 40),anti_aliasing=True)
print(image_224.shape)
plt.imshow(image_224);
```

```python
print(image_40.shape)
plt.imshow(image_40);
```

```python
img_PIL = Image.open(os.path.join(path0, path_data, img_list[0]))
im_224 = img_PIL.resize((224, 224))
display(im_224)
im_224.save(os.path.join(path0, path_resample_224, img_list[0]))
```

```python
%%time
for path in img_list:
    img_PIL = Image.open(os.path.join(path0, path_data, path))
    im_224 = img_PIL.resize((224, 224))
    im_224 = im_224.convert('RGB')
    im_224.save(os.path.join(path0, path_resample_224, path))
    im_40 = img_PIL.resize((40, 40))
    im_40 = im_40.convert('RGB')
    im_40.save(os.path.join(path0, path_resample_40, path))
```
