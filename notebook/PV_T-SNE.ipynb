```python
def path_to_image(path, image_size, num_channels, interpolation):
    img = io_ops.read_file(path)
    img = image_ops.decode_image(img, channels=num_channels, expand_animations=False)
    img = image_ops.resize_images_v2(img, image_size, method=interpolation)
    img.set_shape((image_size[0], image_size[1], num_channels))
    return img
```

```python
def Embedding_generation(root_path, embedding_model):
    img_list = []
    label_list = []
    for root, dirs, files in os.walk(root_path, topdown=False):
        for name in files:
            img_list.append(os.path.join(root, name))
            label_list.append(name)

    DF_list = []
    thumbnails = []
    n_mon = 0
    for im_path, label in zip(img_list, label_list):
        n_mon+=1
        if n_mon%100==0:
            print(f"{n_mon}/{len(label_list)}")
            
        im_iops = path_to_image(im_path, IMG_SIZE, 3, 'bilinear')
        pic_shape = all_data_info.loc[all_data_info['new_filename']==label, ['pixelsy', 'pixelsx']].values[0]
        thumbnails.append(np.round(path_to_image(im_path, tuple(np.uint8(np.round(pic_shape/pic_shape.max()*50))), 3, 'bilinear')))
        im_iops = np.expand_dims(im_iops,axis =0)

        image_embedding = embedding_model.predict(im_iops)
        DF_list.append(pd.DataFrame(image_embedding, index=[label]))
        
        
    Embeddings_df = pd.concat(DF_list)
    return Embeddings_df, img_list, label_list, thumbnails
```

```python
def plot_embedding(X):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    fig, ax = plt.subplots(figsize = (20, 20))
    ax = plt.subplot(111)
    if hasattr(offsetbox, 'AnnotationBbox'):
        # only print thumbnails with matplotlib > 1.0
        shown_images = np.array([[1., 1.]])  # just something big
        for i in range(X.shape[0]):
            dist = np.sum((X[i] - shown_images) ** 2, 1)
            if np.min(dist) < 5e-4:
                # don't show points that are too close
                continue
            #art_id = int(all_data_info2.loc[all_data_info2['new_filename']==label_list[i], 'artist_number'].values[0][1:])
            shown_images = np.r_[shown_images, [X[i]]]
            imagebox = offsetbox.AnnotationBbox( offsetbox.OffsetImage(thumbnails[i]/255), X[i], frameon= False)
                                                #frameon=True, bboxprops =dict(edgecolor=turbo_colormap_data2[art_id]))
            ax.add_artist(imagebox)
    plt.xticks([]), plt.yticks([])
```
